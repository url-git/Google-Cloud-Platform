**Cloud Filestore** to zarządzany serwis plikowy w Google Cloud Platform, który oferuje w pełni zarządzaną usługę przechowywania plików sieciowych, opartą na protokole Network File System (NFS) v3 i v4. Jedną z funkcjonalności Cloud Filestore jest możliwość udostępnienia plików NFS dla instancji wirtualnych maszyn (VM) w chmurze. Dzięki temu aplikacje uruchomione na VM mogą używać plików z Cloud Filestore, jakby były one skonfigurowane lokalnie na tych maszynach.

Udostępnienie plików NFS z Cloud Filestore umożliwia w pełni zarządzane i skalowalne przechowywanie danych dla aplikacji, które wymagają wspólnego dostępu do tych samych plików. Przykładami zastosowań mogą być takie aplikacje jak repozytoria kodu źródłowego, systemy plików udostępniane dla klastrów kontenerów, serwery plików, a także różnego rodzaju narzędzia służące do analizy danych, które wymagają szybkiego dostępu do dużej ilości plików.

**Cloud Filestore** to usługa chmury GCP zapewniająca zarządzane usługi przechowywania plików dla systemów plików NFS (Network File System). Dane NAS (Network-Attached Storage) w kontekście Cloud Filestore to po prostu pliki i foldery, które są przechowywane na obsługiwanym przez usługę dysku sieciowym i są udostępniane przez protokół NFS dla innych zasobów sieciowych. Przechowywane dane mogą być używane przez wiele wirtualnych maszyn, a ich zarządzanie i skalowanie jest łatwe i elastyczne.
